```python
import requests
import io
from bs4 import BeautifulSoup
import pandas as pd
urls = ['https://fbref.com/en/squads/a2d435b3/Leicester-City', 'https://fbref.com/en/squads/47c64c55/Crystal-Palace', 'https://fbref.com/en/squads/822bd0ba/Liverpool', 'https://fbref.com/en/squads/b8fd03ef/Manchester-City', 'https://fbref.com/en/squads/cff3d9bb/Chelsea', 'https://fbref.com/en/squads/19538871/Manchester-United', 'https://fbref.com/en/squads/8cec06e1/Wolverhampton-Wanderers', 'https://fbref.com/en/squads/1df6b87e/Sheffield-United', 'https://fbref.com/en/squads/361ca564/Tottenham-Hotspur', 'https://fbref.com/en/squads/18bb7c10/Arsenal', 'https://fbref.com/en/squads/943e8050/Burnley', 'https://fbref.com/en/squads/d3fd31cc/Everton', 'https://fbref.com/en/squads/b2b47a98/Newcastle-United', 'https://fbref.com/en/squads/33c895d4/Southampton', 'https://fbref.com/en/squads/d07537b9/Brighton--Hove-Albion', 'https://fbref.com/en/squads/7c21e445/West-Ham-United', 'https://fbref.com/en/squads/2abfe087/Watford', 'https://fbref.com/en/squads/4ba7cbea/Bournemouth', 'https://fbref.com/en/squads/8602292d/Aston-Villa', 'https://fbref.com/en/squads/1c781004/Norwich-City']

```

```python
# in every url the team name starts at 37th position of the string
```

```python
#printing all the team names
for url in urls:
    print(url[37:])

#crating a dictionary for the shortnames for each team    
shortname = {
    'Leicester-City' : 'LEI',
    'Crystal-Palace' : 'CRY',
    'Liverpool' : "LIV",
    'Manchester-City' : 'MCI',
    'Chelsea' : 'CHE',
    'Manchester-United' : 'MNU',
    'Wolverhampton-Wanderers' : 'WOL',
    'Sheffield-United' : 'SHU',
    'Tottenham-Hotspur' : 'TOT',
    'Arsenal' : 'ARS',
    'Burnley' : 'BUR',
    'Everton' : 'EVE',
    'Newcastle-United' : 'NEW',
    'Southampton' : 'SOU',
    'Brighton--Hove-Albion' :'BRI',
    'West-Ham-United' : 'WHU',
    'Watford' : 'WAT',
    'Bournemouth' : 'BOU',
    'Aston-Villa' : 'AST',
    'Norwich-City' : 'NOR'
  
}
```

```python
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find_all('table', id = 'ks_sched_3232')
    table = table[0]
    csv_filename = shortname[url[37:]] + '.csv'
    
    with io.open ('stats.txt', 'w', encoding="utf-8") as r:
        r.write('Day,Date,Time,Venue,Result,GF,GA,Opponent,xG,xGA,Attendance,Attendance2,Captain,Formation,Referee,Match Report,,')
        r.write('\n')
                    
        for row in table.find_all('tr'):
            for cell in row.find_all('td'):
                r.write(cell.text)
                r.write(',')
            r.write('\n')
        
    
    df = pd.read_csv('stats.txt')
    df.to_csv('stats.csv', index=None)
    
    df = pd.read_csv('stats.csv')
    df = df.drop(['Match Report','Unnamed: 16', 'Unnamed: 17'], axis= 1)
    df['Attendance']  = df['Attendance']*1000 + (df['Attendance2'])
    df = df.drop('Attendance2', axis = 1)
    df = df.dropna()
    df.to_csv('teamdata/'+csv_filename, index=None)
    
    
print('done')
    
    
    
    
    
    
```
